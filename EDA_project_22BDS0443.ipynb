## The features in this dataset:
1. age (numeric)
2. job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')
3. marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)
4. education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')
5. default: has credit in default? (categorical: 'no','yes','unknown')
6. housing: has housing loan? (categorical: 'no','yes','unknown')
7. loan: has personal loan? (categorical: 'no','yes','unknown')
8. contact: contact communication type (categorical: 'cellular','telephone')
9. month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')
10. day_of_week: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')
11. duration: last contact duration, in seconds (numeric).
12. campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
13. pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
14. previous: number of contacts performed before this campaign and for this client (numeric)
15. poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')
16. emp.var.rate : Employment variation rate (quarterly indicator, numeric).
17. cons.price.idx :Consumer price index (monthly indicator, numeric).
18. cons.conf.idx :Consumer confidence index (monthly indicator, numeric).
19. euribor3m :Euro Interbank Offered Rate for 3 months (numeric).
20. nr.employed : Number of employees (quarterly indicator, numeric).
Output variable (desired target):
21. y. has the client subscribed a term deposit? (binary: 'yes','no')

 ## Reshaping
 A common practice in EDA is converting data between wide and long formats. In the wide format each row represents a record or instance whereas in the long format each row represents a single data point (commonly used in time series)

```python
#for the purpose of reshaping, let's consider a small subset of data.
df_small = df.head(10)
df_small
```

```python
#wide to long
df_small =df_small.stack()
df_small
```

```python
#Unstack the data back to wide format
df_small = df_small.unstack()
df_small
```

## Data Type Optimisation
Categorical data takes up lesser space in memory compared to string data and many Machine Learning algorithms work only with categorical data. Let us see how we can convert the data into category.

```python
df_small['marital']
```

```python
df_small['marital'] = df_small['marital'].astype('category')
print(df_small['marital'])

#check the datatype
print(df['marital'].dtype)
```

## Encode categorical variables
Encoding categorical labels is an important step as most models only work with numerical data. Here, let us encode marital and contact columns. To view what values these columns contain-

```python
print(df_small['marital'].value_counts())
print(df_small['contact'].value_counts())
```

```python
#we will use LabelEncoder
from sklearn.preprocessing import LabelEncoder
le_marital = LabelEncoder()
le_contact = LabelEncoder()
df_small['marital'] = le_marital.fit_transform(df_small['marital'])
df_small['contact'] = le_contact.fit_transform(df_small['contact'])
```

```python
#view the transformations.
print(df_small.head())
```

## Data Smoothing
Data Smoothing involves techniques such as binning.

```python
#we will consider age column of our dataframe for binning purposes.
df_small['age']
```

```python
#equal frequency binning using pd.cut
age_bins = pd.cut(df_small['age'],bins=5)
age_bins
```

```python
#we can also create our custom bins.
print(min(df_small['age']))
print(max(df_small['age']))
```

```python
bins = [24,30,40,50,60]
labels= ['twenties','thirties','forties','fifties']
age_bins = pd.cut(df_small['age'],bins=bins, labels=labels)
df_small['age_cateogry']= age_bins
df_small
```

```python
#also we can distribute using quartiles.
age_bins = pd.qcut(df_small['age'],q=4)
print(age_bins)
```

##Normalisation
 Normalising data is an important step so later on in the model building phase more importance is not assigned to any particular feature (bias).

 ### Min Max Scaler
 The Min Max Scaler works using this formula:

 Values range from 0 to 1 after normalising.

 $$
     X' = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
  $$

```python
df['nr.employed'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df['nr.employed'] = sc.fit_transform(df[['nr.employed']])
df['nr.employed']
```

## Different types of analysis

### Univariate Analysis


*   This type of analysis deals with only one variable at a time and not the relationship between variables.
*   Measures of central tendency (mean, median, mode) find where are values clustering.
* Measures of dispersion such as standard deviation and variance analyse the spread of data.
* Skewness and kurtosis tell us what the graph looks like (Right-tail or left-tail)



The consumer price index variable in the dataset given as 'cons.price.idx' represents how much the average price of some good and services change over time.
